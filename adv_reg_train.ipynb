```python
df[['BsmtQual','BsmtExposure']][df.BsmtExposure.isnull()]
```

```python
df.BsmtExposure = df.BsmtExposure.fillna('NA')
df.BsmtFinType1 = df.BsmtFinType1.fillna('NA')
df.BsmtFinType2 = df.BsmtFinType2.fillna('NA')
df.drop([948],axis=0,inplace=True)
```

```python
df.GarageType = df.GarageType.fillna('NA')
df.GarageYrBlt = df.GarageYrBlt.fillna('NA')
df.GarageFinish = df.GarageFinish.fillna('NA')
df.GarageQual = df.GarageQual.fillna('NA')
df.GarageCond = df.GarageCond.fillna('NA')
```

### Adding test_set to have all possible values for dummy-encoding

```python
X_train = df.drop(['SalePrice'],axis=1)
X_train.shape
```

```python
X_train[1457:]
```

```python
X_train.to_csv('X_train_clean.csv',index=False)
```

```python
test_df = pd.read_csv('test_clean.csv')
test_df.shape
```

```python
test_df.head()
```

```python
final_X = pd.concat([X_train,test_df],axis=0)
final_X.shape
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
final_X[
    [
        'LotFrontage','LotArea','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF',
        '1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','GarageArea','WoodDeckSF','OpenPorchSF',
        'EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal'
    ]
] = scaler.fit_transform(
    final_X[[
        'LotFrontage','LotArea','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF',
        '1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','GarageArea','WoodDeckSF','OpenPorchSF',
        'EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal'
    ]]
)
```

### Dummy Encoding

```python
dummy_train = pd.get_dummies(final_X,drop_first=True)
dummy_train.shape
```

### Get the original train_set

```python
final_X = dummy_train[:1458]
final_X.shape
```

```python
final_X[1457:]
```

### Training

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df.SalePrice
y.head()
```

```python
y.shape
```

### Linear Regression Model

```python
lr_model = LinearRegression()
lr_model.fit(final_X,y)
```

### Random Forrest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
rscv_rfr = RandomizedSearchCV(
    RandomForestRegressor(),
    {
        'criterion':['mse','mae'],
        'max_features' : ["auto", "sqrt", "log2"]        
    },
    cv=5,
    n_iter=6,
    return_train_score=False
)
```

```python
rscv_rfr.fit(final_X,y)
```

```python
pd.DataFrame(rscv_rfr.cv_results_)
```

### XGBoost Regressor

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
simple_xgb = XGBRegressor()
simple_xgb.fit(final_X,y)`
```

```python
rscv_xgb = GridSearchCV(
    XGBRegressor(),
    {
        'max_depth':[10,15,20],
        'min_child_weight':[4,6,8],
        'subsample':[0.8,0.9,1],
        'col_sample_bytree':[0.8,0.9,1],
        'learning_rate':[0.001,0.01,0.1,0.2,0.3]        
    },
    cv=5,
    return_train_score=False
)
```

```python
rscv_xgb.fit(final_X,y)
```

```python
rscv_xgb.best_params_
```

```python
rscv_xgb.best_score_
```
